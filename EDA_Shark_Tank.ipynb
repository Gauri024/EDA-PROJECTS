```python
#saving the file as a dataframe
import pandas as pd
import io
import numpy as mp
import seaborn as sns
import matplotlib.pyplot as plt


data= pd.read_excel(io.BytesIO(uploaded['EDA_Shark_Tank_India.xlsx']))
data.head()
```

**Column Descriptions**

1. episode_number – The episode number in which the pitch was featured.
2. pitch_number – The sequence number of the pitch within the episode.
3. brand_name – The name of the business or startup being pitched.
4. idea – A short description of the business idea or product.
5. deal – A binary indicator (1 = Deal was made, 0 = No deal).
6. pitcher_ask_amount – The amount (in lakhs) the pitcher requested from the sharks.
7. ask_equity – The percentage of equity the pitcher was willing to give in exchange for the asked amount.
8. ask_valuation – The valuation of the company based on the ask amount and equity (%). (Calculated as: pitcher_ask_amount ÷ ask_equity × 100)
9. deal_amount – The final investment amount (in lakhs) agreed upon in the deal.
10. deal_equity – The percentage of equity given to the sharks in the finalized deal.
11. deal_valuation – The valuation of the company based on the final deal. (Calculated as: deal_amount ÷ deal_equity × 100)

**Sharks' Presence in the Panel**
These columns indicate whether a particular shark was present in the panel during the pitch (1 = Present, 0 = Not present):
12. ashneer_present
13. anupam_present
14. aman_present
15. namita_present
16. vineeta_present
17. peyush_present
18. ghazal_present

**Sharks' Investment Decision**
These columns indicate whether a particular shark invested in the deal (1 = Invested, 0 = Did not invest):
19. ashneer_deal
20. anupam_deal
21. aman_deal
22. namita_deal
23. vineeta_deal
24. peyush_deal
25. ghazal_deal

**Investment Distribution**
26. total_sharks_invested – The total number of sharks who invested in the pitch.
27. amount_per_shark – The average amount (in lakhs) each investing shark contributed. (Calculated as: deal_amount ÷ total_sharks_invested)
28. equity_per_shark – The average equity (%) each investing shark received. (Calculated as: deal_equity ÷ total_sharks_invested)

```python
#Which shark has the highest total amount invested?
shark_deal_cols = ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']
for i in shark_deal_cols:
    # Check if any value in the shark_deal column is equal to 1
    if (data[i] == 1).any():
        #Filter the DataFrame to include only the rows where the current shark invested, then sum the deal amount
        total_investment = data[data[i] == 1]['deal_amount'].sum()
        print(f"Total investment by {i.split('_')[0]}: {total_investment}")
```

```python
# Calculate total investment by each shark
shark_deal_cols = ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']
shark_investments = {}
for col in shark_deal_cols:
    shark_name = col.split('_')[0]
    shark_investments[shark_name] = data[data[col] == 1]['deal_amount'].sum()

# Create the pie chart
plt.pie(shark_investments.values(), labels=shark_investments.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Investment Distribution by Sharks')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

From above chart we can see that Aman has invested the most- 20.2%, whereas Ghazal has investe dthe least- 4.7%
